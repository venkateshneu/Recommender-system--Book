```python
books_with_ratings = rating.merge(df, on = "ISBN")
books_with_ratings
```

```python
rating_no = books_with_ratings.groupby("book")["rating"].count().reset_index().rename(columns = {"rating":"ratings_count"})
```

```python
rating_no
```

```python
books_with_ratings
```

```python
total_ratings = books_with_ratings.merge(rating_no, on = "book")
```

```python
total_ratings = total_ratings[total_ratings["ratings_count"]>=50]
```

```python
total_ratings.drop_duplicates( ["user_id", "book"], inplace = True)
```

```python
total_ratings
```

```python
 final_table = total_ratings.pivot_table(columns = "user_id", index = "book", values = "rating")
```

```python
final_table= final_table.fillna(0)
```

```python
final_table
```

```python
from scipy.sparse import csr_matrix
book_matrix = csr_matrix(final_table)
```

```python
type(book_matrix)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm="brute")
```

```python
model.fit(book_matrix)
```

```python
final_table.index
```

```python
distances, suggestions = model.kneighbors(final_table.iloc[155, : ].values.reshape(1, -1), n_neighbors=5)
```
