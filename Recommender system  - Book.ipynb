```python
books_with_ratings = rating.merge(df, on = "ISBN")
books_with_ratings
```

```python
rating_no = books_with_ratings.groupby("book")["rating"].count().reset_index().rename(columns = {"rating":"ratings_count"})
```

```python
rating_no
```

```python
books_with_ratings
```

```python
total_ratings = books_with_ratings.merge(rating_no, on = "book")
```

```python
total_ratings = total_ratings[total_ratings["ratings_count"]>=50]
```

```python
total_ratings.drop_duplicates( ["user_id", "book"], inplace = True)
```

```python
total_ratings
```

```python
 final_table = total_ratings.pivot_table(columns = "user_id", index = "book", values = "rating")
```

```python
final_table= final_table.fillna(0)
```

```python
final_table
```

```python
from scipy.sparse import csr_matrix
book_matrix = csr_matrix(final_table)
```

```python
type(book_matrix)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm="brute")
```

```python
model.fit(book_matrix)
```

```python
final_table.index
```

```python
distances, suggestions = model.kneighbors(final_table.iloc[155, : ].values.reshape(1, -1), n_neighbors=5)
```

```python
 suggestions
```

```python
for i in range(len(suggestions)):
    print(final_table.index[suggestions[i]])
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
normalized_data = scaler.fit_transform(final_table)

kmeans = KMeans(n_clusters=5)
kmeans.fit(normalized_data)


cluster_labels = kmeans.labels_


query_index = 155
query_data = normalized_data[query_index].reshape(1, -1)

query_cluster = kmeans.predict(query_data)[0]


cluster_indices = [i for i, label in enumerate(cluster_labels) if label == query_cluster]


print(final_table.index[cluster_indices])
```

```python
import matplotlib.pyplot as plt


wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(normalized_data)
    wcss.append(kmeans.inertia_)


plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score


silhouette_avg = silhouette_score(normalized_data, cluster_labels)

print("Silhouette Score:", silhouette_avg)
```
